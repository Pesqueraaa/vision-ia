```python
# PROMPT 1: Entorno, datos y trazabilidad inicial

# Importamos y semillas
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns

import os
import json
import hashlib
import sys
import time
from datetime import datetime

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)

# La carga y preparacion de datos
(x_train_full, y_train_full), (x_test, y_test) = keras.datasets.cifar10.load_data()

print(" Formas Iniciales ")
print(f"x_train_full: {x_train_full.shape}")
print(f"y_train_full: {y_train_full.shape}")
print(f"x_test: {x_test.shape}")
print(f"y_test: {y_test.shape}")

x_train, x_valid, y_train, y_valid = train_test_split(
    x_train_full,
    y_train_full,
    test_size=0.2,
    random_state=SEED,
    stratify=y_train_full
)

print("\n Formas Tras Split 80/20 ")
print(f"x_train: {x_train.shape}")
print(f"y_train: {y_train.shape}")
print(f"x_valid: {x_valid.shape}")
print(f"y_valid: {y_valid.shape}")

x_train = x_train.astype('float32') / 255.0
x_valid = x_valid.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

y_train_oh = to_categorical(y_train, 10)
y_valid_oh = to_categorical(y_valid, 10)
y_test_oh = to_categorical(y_test, 10)

print("\n Formas Finales de One-Hot")
print(f"y_train_oh: {y_train_oh.shape}")
print(f"y_valid_oh: {y_valid_oh.shape}")
print(f"y_test_oh: {y_test_oh.shape}")

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10, 10))
for i in range(16):
    plt.subplot(4, 4, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i])
    plt.xlabel(class_names[y_train[i][0]])
plt.show()

# Aplicamos la trazabilidad
env_md_content = f"""
# R2: Environment Versions
- Python: {sys.version.split()[0]}
- TensorFlow: {tf.__version__}
- Keras: {keras.__version__}
- NumPy: {np.__version__}
- Scikit-learn: {sklearn.__version__}
"""
print(env_md_content)

first_1024_bytes = x_train[:1024].tobytes()
data_hash = hashlib.sha256(first_1024_bytes).hexdigest()

data_meta = {
    "timestamp": datetime.now().isoformat(),
    "data_hash_sha256_train_1024": data_hash,
    "shapes": {
        "x_train": list(x_train.shape),
        "y_train_oh": list(y_train_oh.shape),
        "x_valid": list(x_valid.shape),
        "y_valid_oh": list(y_valid_oh.shape),
        "x_test": list(x_test.shape),
        "y_test_oh": list(y_test_oh.shape)
    }
}
print("\n R4: Data Meta (results/data_meta.json) ")
print(json.dumps(data_meta, indent=4))

params = {
    "seed": SEED,
    "data_split": 0.2,
    "normalization": "divide_255",
    "labels": "one_hot",
    "experiments": {}
}
print("\n R3: Params (results/params.json) ")
print(json.dumps(params, indent=4))
```

```python
# PROMPT 2: Baseline MLP (control) + logging

# Importamos
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import os
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import json


#Funciones auxiliares de trazabilidad

def get_run_name():
    return "local_run"

def plot_history(history, model_name):
    if not os.path.exists('figuras'):
        os.makedirs('figuras')
        print("Directorio 'figuras/' creado.")

    df = pd.DataFrame(history.history)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    df[['loss', 'val_loss']].plot(ax=ax1, grid=True)
    ax1.set_title(f"{model_name} - Loss")
    ax1.set_xlabel("Epochs")
    ax1.set_ylabel("Loss")

    df[['accuracy', 'val_accuracy']].plot(ax=ax2, grid=True)
    ax2.set_title(f"{model_name} - Accuracy")
    ax2.set_xlabel("Epochs")
    ax2.set_ylabel("Accuracy")

    plt.tight_layout()
    timestamp = datetime.now().strftime("%Y%m%d_%H%M")
    fig_path = f"figuras/{timestamp}_{model_name}_curves.png"
    plt.savefig(fig_path)
    print(f"Figuras guardadas en {fig_path}")

    plt.show()
    return fig_path

if 'metrics_data' not in globals():
    metrics_data = {}

def save_experiment_results(model_name, history, model, test_loss, test_acc, params_dict):
    if not os.path.exists('results'):
        os.makedirs('results')
        print("Directorio 'results/' creado.")

    run_name = get_run_name()

    print(f" R3: Actualizando params.json para {model_name} ")
    params['experiments'][model_name] = {
        "timestamp": datetime.now().isoformat(),
        "run_name": run_name,
        "params": params_dict
    }
    with open("results/params.json", "w") as f:
        json.dump(params, f, indent=4)

    history_path = f"results/history_{model_name}_{run_name}.csv"
    pd.DataFrame(history.history).to_csv(history_path)
    print(f"History guardado en {history_path}")

    metrics_path = "results/metrics.json"
    new_metrics = {
        model_name: {
            "run_name": run_name,
            "val_acc": max(history.history['val_accuracy']),
            "val_loss": min(history.history['val_loss']),
            "test_acc": test_acc,
            "test_loss": test_loss,
            "total_params": model.count_params()
        }
    }
    metrics_data.update(new_metrics)
    with open(metrics_path, "w") as f:
        json.dump(metrics_data, f, indent=4)

    print(f"\n Métricas (results/metrics.json) ")
    print(json.dumps(metrics_data, indent=4))

# Definimmos el MLP
MODEL_NAME = "mlp_baseline"
INPUT_SHAPE = (32, 32, 3)
NUM_CLASSES = 10
BATCH_SIZE = 64
EPOCHS = 10
LR = 1e-3

mlp_model = keras.Sequential([
    layers.Input(shape=INPUT_SHAPE),
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(NUM_CLASSES, activation='softmax')
], name=MODEL_NAME)

mlp_model.summary()
print(f"Total params (anotado): {mlp_model.count_params()}")

# Compilamos y entrenamos
mlp_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=LR),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_mlp = mlp_model.fit(
    x_train, y_train_oh,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_data=(x_valid, y_valid_oh),
    verbose=1
)

# Creamos la grafica y evaluamos
fig_path_mlp = plot_history(history_mlp, MODEL_NAME)

print("\n--- Evaluación (Test Set) ---")
test_loss_mlp, test_acc_mlp = mlp_model.evaluate(x_test, y_test_oh, verbose=0)
print(f"MLP Test Loss: {test_loss_mlp:.4f}")
print(f"MLP Test Accuracy: {test_acc_mlp:.4f}")

# Trazabilidad
mlp_params = {
    "model_type": "MLP",
    "layers": [
        {"type": "Flatten"},
        {"type": "Dense", "units": 256, "activation": "relu"},
        {"type": "Dropout", "rate": 0.5},
        {"type": "Dense", "units": 10, "activation": "softmax"}
    ],
    "optimizer": "Adam",
    "learning_rate": LR,
    "batch_size": BATCH_SIZE,
    "epochs": EPOCHS
}

save_experiment_results(MODEL_NAME, history_mlp, mlp_model, test_loss_mlp, test_acc_mlp, mlp_params)
```

```python
# PROMPT 3: CNN simple de 2 bloques y logging

# Importaciones
from tensorflow import keras
from tensorflow.keras import layers
import time


# Definimos las constantes necesarias
INPUT_SHAPE = (32, 32, 3)
NUM_CLASSES = 10
BATCH_SIZE = 64
LR = 1e-3

# Definimos la CNN
MODEL_NAME_CNN = "cnn_2_block"
EPOCHS_CNN = 15

def build_cnn_2_block(input_shape=(32, 32, 3), num_classes=10):
    model = keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', name='conv1_1'),
        layers.MaxPooling2D((2, 2), name='pool1'),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same', name='conv2_1'),
        layers.MaxPooling2D((2, 2), name='pool2'),
        layers.Flatten(name='flatten'),
        layers.Dense(128, activation='relu', name='dense1'),
        layers.Dropout(0.5, name='dropout'),
        layers.Dense(num_classes, activation='softmax', name='output')
    ], name=MODEL_NAME_CNN)
    return model

cnn_model_2b = build_cnn_2_block(INPUT_SHAPE, NUM_CLASSES)
cnn_model_2b.summary()
print(f"Total params (anotado): {cnn_model_2b.count_params()}")

# Compilamos y entrenamos
cnn_model_2b.compile(
    optimizer=keras.optimizers.Adam(learning_rate=LR),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

start_time = time.time()
history_cnn_2b = cnn_model_2b.fit(
    x_train, y_train_oh,
    epochs=EPOCHS_CNN,
    batch_size=BATCH_SIZE,
    validation_data=(x_valid, y_valid_oh),
    verbose=1
)
end_time = time.time()

time_per_epoch_cnn_2b = (end_time - start_time) / EPOCHS_CNN
print(f"Tiempo promedio por época (CNN 2-Block): {time_per_epoch_cnn_2b:.2f}s")

#Creamos la grafica y evaluamos
fig_path_cnn_2b = plot_history(history_cnn_2b, MODEL_NAME_CNN)

print("\n--- Evaluación (Test Set) ---")
test_loss_cnn_2b, test_acc_cnn_2b = cnn_model_2b.evaluate(x_test, y_test_oh, verbose=0)
print(f"CNN 2-Block Test Loss: {test_loss_cnn_2b:.4f}")
print(f"CNN 2-Block Test Accuracy: {test_acc_cnn_2b:.4f}")

# Trazabilidad
cnn_2b_params = {
    "model_type": "CNN",
    "blocks": 2,
    "filters": [32, 64],
    "dense_units": 128,
    "dropout": 0.5,
    "optimizer": "Adam",
    "learning_rate": LR,
    "batch_size": BATCH_SIZE,
    "epochs": EPOCHS_CNN,
    "time_per_epoch_sec": time_per_epoch_cnn_2b
}

save_experiment_results(MODEL_NAME_CNN, history_cnn_2b, cnn_model_2b, test_loss_cnn_2b, test_acc_cnn_2b, cnn_2b_params)

# Creamos una tabla comparativa
print("\n--- TABLA COMPARATIVA (MLP vs CNN 2-Block) ---")
print(f"| Modelo       | Parámetros | Test Acc | Val Acc (Max) |")
print(f"|--------------|------------|----------|---------------|")
print(f"| MLP Baseline | {metrics_data['mlp_baseline']['total_params']:<10} | {metrics_data['mlp_baseline']['test_acc']:.4f}   | {metrics_data['mlp_baseline']['val_acc']:.4f}         |")
print(f"| CNN 2-Block  | {metrics_data['cnn_2_block']['total_params']:<10} | {metrics_data['cnn_2_block']['test_acc']:.4f}   | {metrics_data['cnn_2_block']['val_acc']:.4f}         |")
```

```python
# PROMPT 4: L2 y EarlyStopping y el logging

# Importaciones
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping
import time

# Constantes
INPUT_SHAPE = (32, 32, 3)
NUM_CLASSES = 10
BATCH_SIZE = 64
LR = 1e-3

# Definimos el CNN con L2
MODEL_NAME_CNN_L2 = "cnn_2_block_L2_ES"
EPOCHS_CNN_L2 = 30
L2_RATE = 1e-4

def build_cnn_2_block_l2(input_shape=(32, 32, 3), num_classes=10, l2_rate=1e-4):
    l2_reg = keras.regularizers.l2(l2_rate)

    model = keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same',
                      kernel_regularizer=l2_reg, name='conv1_1'),
        layers.MaxPooling2D((2, 2), name='pool1'),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same',
                      kernel_regularizer=l2_reg, name='conv2_1'),
        layers.MaxPooling2D((2, 2), name='pool2'),
        layers.Flatten(name='flatten'),
        layers.Dense(128, activation='relu',
                     kernel_regularizer=l2_reg, name='dense1'),
        layers.Dropout(0.5, name='dropout'),
        layers.Dense(num_classes, activation='softmax', name='output')
    ], name=MODEL_NAME_CNN_L2)
    return model

cnn_model_l2 = build_cnn_2_block_l2(INPUT_SHAPE, NUM_CLASSES, L2_RATE)
cnn_model_l2.summary()

# Creamos los callbacks y compilamos
early_stopping_cb = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

cnn_model_l2.compile(
    optimizer=keras.optimizers.Adam(learning_rate=LR),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Entrenamos
start_time = time.time()
history_cnn_l2 = cnn_model_l2.fit(
    x_train, y_train_oh,
    epochs=EPOCHS_CNN_L2,
    batch_size=BATCH_SIZE,
    validation_data=(x_valid, y_valid_oh),
    callbacks=[early_stopping_cb],
    verbose=1
)
end_time = time.time()

epochs_run_l2 = len(history_cnn_l2.epoch)
time_per_epoch_cnn_l2 = (end_time - start_time) / epochs_run_l2
print(f"Tiempo promedio por época (CNN L2): {time_per_epoch_cnn_l2:.2f}s")
if early_stopping_cb.stopped_epoch > 0:
    print(f"Entrenamiento detenido en época: {early_stopping_cb.stopped_epoch}")

# Creamos la grafica y evaluamos
fig_path_cnn_l2 = plot_history(history_cnn_l2, MODEL_NAME_CNN_L2)

print("\n--- Evaluación (Test Set) ---")
test_loss_cnn_l2, test_acc_cnn_l2 = cnn_model_l2.evaluate(x_test, y_test_oh, verbose=0)
print(f"CNN L2+ES Test Loss: {test_loss_cnn_l2:.4f}")
print(f"CNN L2+ES Test Accuracy: {test_acc_cnn_l2:.4f}")

# Trazabilidad
cnn_l2_params = {
    "model_type": "CNN",
    "blocks": 2,
    "filters": [32, 64],
    "dense_units": 128,
    "dropout": 0.5,
    "regularization": {"L2": L2_RATE},
    "callbacks": {"EarlyStopping": {"monitor": "val_loss", "patience": 5}},
    "optimizer": "Adam",
    "learning_rate": LR,
    "batch_size": BATCH_SIZE,
    "epochs_trained": epochs_run_l2,
    "epochs_max": EPOCHS_CNN_L2,
    "time_per_epoch_sec": time_per_epoch_cnn_l2
}

save_experiment_results(MODEL_NAME_CNN_L2, history_cnn_l2, cnn_model_l2, test_loss_cnn_l2, test_acc_cnn_l2, cnn_l2_params)

# Actualizamos la tabla anterior
print("\n TABLA COMPARATIVA ")
print(f"| Modelo       | Parámetros | Test Acc | Val Acc (Max) |")
print(f"|--------------|------------|----------|---------------|")
print(f"| MLP Baseline | {metrics_data['mlp_baseline']['total_params']:<10} | {metrics_data['mlp_baseline']['test_acc']:.4f}   | {metrics_data['mlp_baseline']['val_acc']:.4f}         |")
print(f"| CNN 2-Block  | {metrics_data['cnn_2_block']['total_params']:<10} | {metrics_data['cnn_2_block']['test_acc']:.4f}   | {metrics_data['cnn_2_block']['val_acc']:.4f}         |")
print(f"| CNN L2+ES    | {metrics_data[MODEL_NAME_CNN_L2]['total_params']:<10} | {metrics_data[MODEL_NAME_CNN_L2]['test_acc']:.4f}   | {metrics_data[MODEL_NAME_CNN_L2]['val_acc']:.4f}         |")
```

```python
# PROMPT 5: Data Augmentation + ReduceLROnPlateau + logging

# Importaciones
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import os
from datetime import datetime
import time

# Definimos las constantes
INPUT_SHAPE = (32, 32, 3)
NUM_CLASSES = 10
BATCH_SIZE = 64
LR = 1e-3
L2_RATE = 1e-4
SEED = 42

# Diseñamos la argumentacion
data_augmentation_layers = keras.Sequential(
    [
        layers.Input(shape=INPUT_SHAPE),
        layers.RandomFlip("horizontal", seed=SEED),
        layers.RandomRotation(0.1, seed=SEED),
        layers.RandomZoom(0.1, seed=SEED),
    ],
    name="data_augmentation"
)

# Visualizamos y analizamos
plt.figure(figsize=(10, 10))
first_image = x_train[6]
for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    augmented_image = data_augmentation_layers(tf.expand_dims(first_image, 0), training=True)
    plt.imshow(augmented_image[0])
    plt.axis("off")

if not os.path.exists('figuras'): os.makedirs('figuras')
timestamp = datetime.now().strftime("%Y%m%d_%H%M")
fig_path_aug = f"figuras/{timestamp}_augmentation_preview.png"
plt.savefig(fig_path_aug)
print(f"Figura de Augmentation guardada en {fig_path_aug}")
plt.show()

# Definimos el modelo
MODEL_NAME_CNN_AUG = "cnn_2_block_Aug_L2_ES_RLROP"
EPOCHS_CNN_AUG = 20

#Definimos el bloque como antes
def build_cnn_2_block_l2(input_shape=(32, 32, 3), num_classes=10, l2_rate=1e-4):
    l2_reg = keras.regularizers.l2(l2_rate)
    model = keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', kernel_regularizer=l2_reg, name='conv1_1'),
        layers.MaxPooling2D((2, 2), name='pool1'),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=l2_reg, name='conv2_1'),
        layers.MaxPooling2D((2, 2), name='pool2'),
        layers.Flatten(name='flatten'),
        layers.Dense(128, activation='relu', kernel_regularizer=l2_reg, name='dense1'),
        layers.Dropout(0.5, name='dropout'),
        layers.Dense(num_classes, activation='softmax', name='output')
    ], name="cnn_2_block_l2_base")
    return model

cnn_model_aug = build_cnn_2_block_l2(INPUT_SHAPE, NUM_CLASSES, L2_RATE)
model_with_aug = keras.Sequential([
    data_augmentation_layers,
    cnn_model_aug
], name=MODEL_NAME_CNN_AUG)

# Callbacks
early_stopping_cb = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)
reduce_lr_cb = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=3,
    min_lr=1e-6,
    verbose=1
)
callbacks_list = [early_stopping_cb, reduce_lr_cb]

#Compilamos y entrenamos
model_with_aug.compile(
    optimizer=keras.optimizers.Adam(learning_rate=LR),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model_with_aug.build(input_shape=(None, 32, 32, 3))
model_with_aug.summary()

start_time = time.time()
history_cnn_aug = model_with_aug.fit(
    x_train, y_train_oh,
    epochs=EPOCHS_CNN_AUG,
    batch_size=BATCH_SIZE,
    validation_data=(x_valid, y_valid_oh),
    callbacks=callbacks_list,
    verbose=1
)
end_time = time.time()

epochs_run_aug = len(history_cnn_aug.epoch)
time_per_epoch_cnn_aug = (end_time - start_time) / epochs_run_aug
print(f"Tiempo promedio por época CNN Aug: {time_per_epoch_cnn_aug:.2f}s")

# Evaluamos y creamos la grafica
fig_path_cnn_aug = plot_history(history_cnn_aug, MODEL_NAME_CNN_AUG)

print("\n Evaluación (Test Set) ")
test_loss_cnn_aug, test_acc_cnn_aug = model_with_aug.evaluate(x_test, y_test_oh, verbose=0)
print(f"CNN Aug+L2+ES+RLROP Test Loss: {test_loss_cnn_aug:.4f}")
print(f"CNN Aug+L2+ES+RLROP Test Accuracy: {test_acc_cnn_aug:.4f}")

# Trzabilidad
cnn_aug_params = {
    "model_type": "CNN",
    "blocks": 2,
    "filters": [32, 64],
    "dense_units": 128,
    "dropout": 0.5,
    "regularization": {"L2": L2_RATE, "DataAugmentation": True},
    "callbacks": {
        "EarlyStopping": {"monitor": "val_loss", "patience": 5},
        "ReduceLROnPlateau": {"monitor": "val_loss", "patience": 3, "factor": 0.2}
    },
    "optimizer": "Adam",
    "learning_rate_initial": LR,
    "batch_size": BATCH_SIZE,
    "epochs_trained": epochs_run_aug,
    "epochs_max": EPOCHS_CNN_AUG,
    "time_per_epoch_sec": time_per_epoch_cnn_aug
}

save_experiment_results(MODEL_NAME_CNN_AUG, history_cnn_aug, model_with_aug, test_loss_cnn_aug, test_acc_cnn_aug, cnn_aug_params)
```

```python
# PROMPT 8: SGD + CosineDecay (alternativa) + logging

# Importaciones
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping
import pandas as pd
import matplotlib.pyplot as plt
import time

# Constantes
INPUT_SHAPE = (32, 32, 3)
NUM_CLASSES = 10
BATCH_SIZE = 64
L2_RATE = 1e-4

# Definimos el CNN
MODEL_NAME_SGD = "cnn_3_block_SGD_Cosine"
EPOCHS_SGD = 30
LR_SGD_INITIAL = 0.05
MOMENTUM_SGD = 0.9

total_steps = (len(x_train) / BATCH_SIZE) * EPOCHS_SGD
try:
    cosine_scheduler = keras.optimizers.schedules.CosineDecay(
        initial_learning_rate=LR_SGD_INITIAL,
        decay_steps=total_steps
    )
except AttributeError:
     cosine_scheduler = keras.experimental.CosineDecay(
        initial_learning_rate=LR_SGD_INITIAL,
        decay_steps=total_steps
    )

# Rempilamos el modelo
if 'build_cnn_3_block_l2' not in globals():
    def build_cnn_3_block_l2(input_shape=(32, 32, 3), num_classes=10, l2_rate=1e-4):
        l2_reg = keras.regularizers.l2(l2_rate)
        model = keras.Sequential([
            layers.Input(shape=input_shape),
            layers.Conv2D(32, (3, 3), activation='relu', padding='same', kernel_regularizer=l2_reg, name='conv1_1'),
            layers.MaxPooling2D((2, 2), name='pool1'),
            layers.Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=l2_reg, name='conv2_1'),
            layers.MaxPooling2D((2, 2), name='pool2'),
            layers.Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=l2_reg, name='conv3_1'),
            layers.MaxPooling2D((2, 2), name='pool3'),
            layers.Flatten(name='flatten'),
            layers.Dense(128, activation='relu', kernel_regularizer=l2_reg, name='dense1'),
            layers.Dropout(0.5, name='dropout'),
            layers.Dense(num_classes, activation='softmax', name='output')
        ], name="cnn_3_block_base")
        return model

cnn_model_3b_base_sgd = build_cnn_3_block_l2(INPUT_SHAPE, NUM_CLASSES, L2_RATE)
model_sgd = keras.Sequential([
    data_augmentation_layers,
    cnn_model_3b_base_sgd
], name=MODEL_NAME_SGD)

optimizer_sgd = keras.optimizers.SGD(
    learning_rate=cosine_scheduler,
    momentum=MOMENTUM_SGD
)
model_sgd.compile(
    optimizer=optimizer_sgd,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model_sgd.build(input_shape=(None, 32, 32, 3))
model_sgd.summary()

# Lo entrenamos
early_stopping_cb_sgd = EarlyStopping(
    monitor='val_loss',
    patience=7,
    restore_best_weights=True,
    verbose=1
)

start_time = time.time()
history_cnn_sgd = model_sgd.fit(
    x_train, y_train_oh,
    epochs=EPOCHS_SGD,
    batch_size=BATCH_SIZE,
    validation_data=(x_valid, y_valid_oh),
    callbacks=[early_stopping_cb_sgd],
    verbose=1
)
end_time = time.time()

epochs_run_sgd = len(history_cnn_sgd.epoch)
time_per_epoch_cnn_sgd = (end_time - start_time) / epochs_run_sgd
print(f"Tiempo promedio por época (CNN SGD): {time_per_epoch_cnn_sgd:.2f}s")

# Creamos la grafica y evaluamos
fig_path_cnn_sgd = plot_history(history_cnn_sgd, MODEL_NAME_SGD)

print("\n--- Evaluación (Test Set) ---")
test_loss_cnn_sgd, test_acc_cnn_sgd = model_sgd.evaluate(x_test, y_test_oh, verbose=0)
print(f"CNN SGD+Cosine Test Loss: {test_loss_cnn_sgd:.4f}")
print(f"CNN SGD+Cosine Test Accuracy: {test_acc_cnn_sgd:.4f}")

# Trazabilidad
cnn_sgd_params = {
    "model_type": "CNN",
    "blocks": 3,
    "filters": [32, 64, 128],
    "dense_units": 128,
    "dropout": 0.5,
    "regularization": {"L2": L2_RATE, "DataAugmentation": True},
    "callbacks": {
        "EarlyStopping": {"monitor": "val_loss", "patience": 7}
    },
    "optimizer": "SGD",
    "momentum": MOMENTUM_SGD,
    "lr_schedule": "CosineDecay",
    "learning_rate_initial": LR_SGD_INITIAL,
    "batch_size": BATCH_SIZE,
    "epochs_trained": epochs_run_sgd,
    "epochs_max": EPOCHS_SGD,
    "time_per_epoch_sec": time_per_epoch_cnn_sgd
}

save_experiment_results(MODEL_NAME_SGD, history_cnn_sgd, model_sgd, test_loss_cnn_sgd, test_acc_cnn_sgd, cnn_sgd_params)

# Comparamos las curvas
df_adam = pd.DataFrame(history_cnn_3b.history)
df_sgd = pd.DataFrame(history_cnn_sgd.history)

plt.figure(figsize=(10, 6))
plt.plot(df_adam['val_accuracy'], label=f'Adam (3B) - {MODEL_NAME_CNN_3B}', color='blue')
plt.plot(df_sgd['val_accuracy'], label=f'SGD+Cosine (3B) - {MODEL_NAME_SGD}', color='orange')
plt.title('Comparativa Optimizadores (Val Accuracy)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
